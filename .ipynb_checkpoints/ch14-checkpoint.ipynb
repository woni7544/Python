```python
# t-검정 : 두 집단의 평균에 통계적으로 유의한 차이가 있는지 알아볼 때 사용하는 통계분석기법
```

```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
```

```python
mpg.query('category in ["compact", "suv"]')\
   .groupby('category', as_index = False)\
   .agg(n = ('category','count'),
        mean = ('cty','mean'))
```

```python
# 평균 차이가 통계적으로 유의한지 t-검정을 통해서 알 수 있다
```

```python
# compact 자동차와 suv 자동차의 도시 연비 t-검정
```

```python
compact = mpg.query('category == "compact"')['cty'] # 'compact'행의 'cty'를 출력
suv = mpg.query('category == "suv"')['cty']
```

```python
# t-test
from scipy import stats
stats.ttest_ind(compact,suv,equal_var = True)
# p-value가 유의확률 : 보통 5%(0.05) 미만일 경우 유의하다고 판단(우연히 일어난 일이 아니라는 뜻)
```

```python
# 일반 휘발유와 고급 휘발유의 도시 연비 t-검정
```

```python
mpg.query('fl in ["r", "p"]')\
    .groupby('fl', as_index = False)\
    .agg(n = ('fl','count'),
         mean = ('cty', 'mean'))
```

```python
regular = mpg.query('fl == "r"')['cty']
premium = mpg.query('fl == "p"')['cty']

# t-test
stats.ttest_ind(regular,premium,equal_var = True)
```

```python
# 실업자 수와 개인 소비 지출의 상관관계
```

```python
economics = pd.read_csv('economics.csv')
```

```python
economics_corr = economics[['unemploy','pce']].corr()
economics_corr
```

```python
# 상관계수와 유의확률 둘 다 확인하기
```

```python
stats.pearsonr(economics['unemploy'], economics['pce'])
```

```python
import seaborn as sns
sns.heatmap(economics_corr,annot = True, cmap = 'RdBu')
```

```python
# 상관행렬 히트맵 만들기
```

```python
mtcars = pd.read_csv('mtcars.csv')
mtcars.head()
```

```python
car_cor = mtcars.corr()# 상관행렬 만들기
car_cor = round(car_cor,2) # 소수 둘째 자리까지 반올림
car_cor
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi'  : '120',
                     'figure.figsize' : [7.5,5.5]})
```

```python
sns.heatmap(car_cor, annot = True, cmap = 'RdBu')
```

```python
# 대각행렬 중복제거
```

```python
import numpy as np
mask = np.zeros_like(car_cor)
mask
```

```python
mask[np.triu_indices_from(mask)] = 1 # 오른쪽 위 대각 행렬을 1로 변경
mask
```

```python
sns.heatmap(car_cor, annot = True, cmap = 'RdBu', mask = mask)
```
