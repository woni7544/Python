```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'sex'  : ['M','F', np.nan, 'M', 'F'],
                   'score': [5, 4, 3, 4, np.nan]})
df
```

```python
print(pd.isna(df)) # 결측치 데이터 'TRUE'로 추출
print(pd.isna(df).sum()) # 결측치 데이터가 몇 개인지 출력
```

```python
#결측치 제거하기
```

```python
#.dropna(subset = []) : subset 안에 있는 행의 결측치가 아닌 값만 출력 
```

```python
df.dropna(subset = ['sex'])
```

```python
df_nomiss = df.dropna(subset = ['score', 'sex'])
df_nomiss
```

```python
df_nomiss2 = df.dropna() # Nan이 하나라도 있으면 제거
df_nomiss2
```

```python
# 결측치 대체하기
# 데이터는 적고 결측치가 많을 경우, 결측치를 제거하면 분석결과가 왜곡될 수 있어서 제거대신 다른 값으로 대체
```

```python
exam = pd.read_csv('exam.csv')
exam.loc[[2,7,14],['math']] = np.nan # 2,7,14행의 math에 NaN 할당
exam
```

```python
#NaN의 값을 math 평균인 55로 대체
```

```python
exam['math'].mean()
```

```python
exam['math'] = exam['math'].fillna(55) #.fillna() : NaN을 다른 값으로 대체
exam
```

```python
exam['math'].isna().sum() #결측치 빈도 확인
```

```python
# 범위 외 이상치 값 제거
```

```python
#.value_counts()
```

```python
df = pd.DataFrame({'sex'  : [1,2,1,3,2,1],
                   'score': [5,4,3,4,2,6]})
df
```

```python
df['sex'].value_counts().sort_index()
```

```python
df['sex'] = np.where(df['sex'] == 3, np.nan, df['sex'])
df
```

```python
df['score'].value_counts().sort_index()
```

```python
df['score'] = np.where(df['score'] > 5, np.nan, df['score'])
df
```

```python
df_miss = df.dropna(subset = ['sex', 'score']) \
            .groupby('sex') \
            .agg(mean_score = ('score', 'mean'))
df_miss
```
