```python
# 파생변수 추가하기
```

```python
exam.assign(total = exam['math'] + exam['english'] + exam['science'],
            mean = (exam['math'] + exam['english'] + exam['science']) / 3)
# assgin() : 데이터에 파생변수 추가 가능 
```

```python
import numpy as np
exam.assign(test = np.where(exam['science'] >= 60, 'pass', 'fail')) # assign 안에 np.where 사용가능
```

```python
# 집단 별 요약하기
```

```python
exam.agg(mean_math = ('math', 'mean')) # agg() : 요약 통계량 구하기
```

```python
exam.groupby('nclass').agg(mean_math = ('math', 'mean')) # groupby: 변수의 범주별로 데이터 분리
```

```python
exam.groupby('nclass', as_index = False).agg(mean_math = ('math', 'mean'))
# as_index = False : 변수를 인덱스로 바꾸지 않고 원래대로 유지
```

```python
# agg()에 자주 사용하는 요약 통계량 함수
# mean(), std() : 표준편차 , sum(), median() : 중앙값, min(), max(), count() : 빈도
```

```python
exam.groupby('nclass', as_index = False).mean() # 반 별 과목 평균점수
```

```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
```

```python
mpg.info()
```

```python
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean'))
# 제조회사 및 구동 방식별 분리
# cty 평균 하위 집단 분리
```

```python
# pandas 함수 조합하기
```

```python
# 제조 회사별로 'suv' 자동차의 도시 및 고속도로 합산 연비 평균을 구해 내림차순으로 정렬하고, 1~5위 출력하기
```

```python
# 'suv' 추출 : query()
# 합산연비 변수 생성 : assign()
# 회사별로 분리 : groupby()
# 합산연비 평균 구하기 : agg()
# 내림차순 정렬 : sort_values()
# 1~5위 출력 : head()
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.query("category == 'suv'").assign(total = (mpg['cty'] + mpg['hwy']) / 2) \
   .groupby('manufacturer').agg( mean_total = ('total', 'mean')) \
   .sort_values('mean_total', ascending = False) \
   .head()
```

```python
# category 별 cty 평균 구하기
```

```python
mpg.groupby('category').agg(mean_cty = ('cty', 'mean'))
```

```python
# cty 평균이 높은 순으로 정렬
```

```python
mpg.groupby('category').agg(mean_cty = ('cty', 'mean')).sort_values('mean_cty', ascending = False)
```

```python
# hwy 평균이 가장 높은 회사 세 곳을 출력
```

```python
mpg.groupby('manufacturer').agg(mean_hwy = ('hwy','mean')) \
   .sort_values('mean_hwy', ascending = False).head(3)
```

```python
# 회사 별 'compact' 차종 수를 내림차순으로 정렬
```

```python
mpg.query("category == 'compact'").groupby('manufacturer').agg(n = ('manufacturer', 'count')) \
   .sort_values('n', ascending = False)
```

```python
# 가로로 데이터 합치기
```

```python
test1 = pd.DataFrame({'id'      : [1,2,3,4,5],
                      'midterm' : [60,80,70,90,85]}) # 중간고사 데이터 생성
test2 = pd.DataFrame({'id'      : [1,2,3,4,5],
                      'final'   : [70,83,65,95,80]}) # 기말고사 데이터 생성
```

```python
test1
```

```python
test2
```

```python
pd.merge(test1, test2, how = 'left', on = 'id')
# merge : 가로 데이터 합병
# how = 'left' : 두번째 입력한 데이터를 왼쪽에 붙이겠다는 뜻
# on : 기준으로 삼을 변수명
```

```python
# 세로로 데이터 합치기
```

```python
group_a = pd.DataFrame({'id'   : [1,2,3,4,5],
                        'test' : [60,80,70,90,85]}) # 1~5번 학생 시험 데이터 생성
group_b = pd.DataFrame({'id'   : [6,7,8,9,10],
                        'test' : [70,83,65,95,80]}) # 6~10번 학생 시험 데이터 생성
```

```python
group_a
```

```python
group_b
```

```python
group_all = pd.concat([group_a, group_b])
group_all
# concat : 세로 데이터 합병
# 인덱스 번호가 중복되는 것을 확인
```

```python
group_all = pd.concat([group_a, group_b], ignore_index = True)
group_all
# ignore_index = True : 인덱스 번호를 새로 부여
```
