```python
import numpy as np
```

```python
import pandas as pd
```

```python
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
mpg.info()
```

```python
mpg.describe()
```

```python
mpg['test'] = np.where(mpg['total'] >= 30, 'pass', 'fail' )
mpg.head()
```

```python
mpg['test'].value_counts() # pass와 fail 빈도표 만들기
```

```python
# 막대그래프로 빈도 표현하기
```

```python
count_test = mpg['test'].value_counts()
```

```python
count_test.plot.bar()
```

```python
count_test.plot.bar(rot = 0) # 축이름 회전 --> rot
```

```python
#중첩 조건문
```

```python
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
               np.where(mpg['total'] >= 20, 'B', 'C'))
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts() # 등급 빈도 먼저ㅓ 만들기
```

```python
count_grade.plot.bar(rot = 0) # 등급 빈도 막대 그래프 생성
```

```python
count_grade
```

```python
count_grade = mpg['grade'].value_counts().sort_index() # 알파벳 순으로 나열 sort_index()
```

```python
midwest = pd.read_csv('midwest.csv')
midwest
```

```python
midwest.info()
```

```python
midwest = midwest.rename(columns = {'poptotal' : 'total'})
midwest = midwest.rename(columns = {'asain' : 'asian'})
midwest # 특정 column 변수명 바꾸기
```

```python
midwest['asian_per'] = midwest['asian'] / midwest['total'] * 100
```

```python
midwest['asian_per'].plot.hist()
```

```python
midwest = midwest.rename( columns = {'asian_per' : 'ratio'})
```

```python
midwest['mean'] = np.where(midwest['ratio'] > sum(midwest['ratio']) / len(midwest), 'Large', 'Small')
```

```python
midwest
```

```python
midwest['mean'].value_counts()
```
